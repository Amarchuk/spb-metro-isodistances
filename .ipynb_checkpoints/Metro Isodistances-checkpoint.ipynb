```python
from BeautifulSoup import BeautifulSoup
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
from scipy.misc import imread
import matplotlib.image as mpimg
from mpl_toolkits.basemap import Basemap, shiftgrid, cm
import requests
import math
import os

%matplotlib inline
```

```python
#SPb center
c_lon, c_lat = 30.3159, 59.9391 
#Image radius in degrees
spn = 0.3
spb_map = Image(filename='spb_small.png')
spb_map
#Use Yandex Static Maps API request for get an image; 600x450 is a max size of static image  
#https://static-maps.yandex.ru/1.x/?ll=30.3159,59.9391&spn=0.3,0.3&size=600,450&l=map
```

```python
#read metro stations list
stations = []
f = open('stations.txt', 'r')
for line in f.readlines():
    stations.append(line)
print len(stations)
```

```python
#Get station longitude and latitude from it name with Yandex Geocode PI  
payload = {'geocode': 'метро Маяковская', 'll': '30.3159,59.93914', 'spn' : '0.552069,0.400552'}

lons, lats = [], []
for station in stations:
    payload['geocode'] = 'метро '+ station
    response = requests.get('https://geocode-maps.yandex.ru/1.x/',
                         params=payload)
    y=BeautifulSoup(response.text)
    results = y.findAll("pos")
    if results:
        lon, lat = str(results[0])[5:-6].split(' ')
        lons.append(float(lon))
        lats.append(float(lat))
```

```python
img = mpimg.imread('spb_small.png')
fig = plt.figure(figsize=[10,10])
plt.scatter([c_lon], [c_lat], color ='red', zorder=1)
plt.scatter(lons, lats, zorder=1)
plt.imshow(img, zorder=0, extent=[c_lon-spn, c_lon+spn, c_lat-spn, c_lat+spn])
plt.show()
```

```python
def distance_on_unit_sphere(lat1, long1, lat2, long2):
    '''rough distance in km'''
    degrees_to_radians = math.pi/180.0
         
    phi1 = (90.0 - lat1)*degrees_to_radians
    phi2 = (90.0 - lat2)*degrees_to_radians
         
    theta1 = long1*degrees_to_radians
    theta2 = long2*degrees_to_radians
     
    cos = (math.sin(phi1)*math.sin(phi2)*math.cos(theta1 - theta2) + 
           math.cos(phi1)*math.cos(phi2))
    arc = math.acos(cos)
    return arc*6373

#create coordinate grid
lon_range = np.linspace(c_lon-spn, c_lon+spn, 500)
lat_range = np.linspace(c_lat-spn, c_lat+spn, 500)

dist = np.empty(250000).reshape(500, 500)
dist.fill(100)
```

```python
#calculate distances
for ind1, lo in np.ndenumerate(lon_range):
    for ind2, la in np.ndenumerate(lat_range):
        for i in range(len(lons)):
            dist[ind2][ind1] = min(distance_on_unit_sphere(la, lo, lats[i], lons[i]), dist[ind2][ind1])
```

```python
fig, axes = plt.subplots(ncols=2, figsize=[18,12])
fig.subplots_adjust(hspace=.2)

def plot_basemap(ax):
    my_map = Basemap(rsphere=(6378137.00, 6356752.3142), projection='lcc', lat_0=c_lat, lon_0=c_lon,
        resolution = 'h', area_thresh = 1.0, ax = ax,
        llcrnrlon=c_lon-spn, llcrnrlat=c_lat-spn,
        urcrnrlon=c_lon+spn, urcrnrlat=c_lat+spn)
    
    x,y = my_map(np.array(lons), np.array(lats))
    my_map.plot(x, y, 's', markersize=2)

    parallels = np.arange(59.,61,0.1)
    my_map.drawparallels(parallels,labels=[False,True,True,False])
    meridians = np.arange(29.,31.,0.1)
    my_map.drawmeridians(meridians,labels=[True,False,False,True])
    return my_map

m1 = plot_basemap(axes[0])
x0, y0 = m1(29.93, 59.783)
x1, y1 = m1(30.70, 60.093)
axes[0].imshow(img,  zorder=0, extent = (x0, x1, y0, y1))
ny = dist.shape[0]; nx = dist.shape[1]
lon_range, lat_range = m1.makegrid(nx, ny)
x, y = m1(lon_range, lat_range)
clevs = [0., 1., 1.5, 2.5, 3., 5, 7.5, 10, 15, 20, 30]
cs = m1.contour(x, y, dist, clevs, linewidths=1.5, alpha=0.5, zorder=1)

m2 = plot_basemap(axes[1])
ny = dist.shape[0]; nx = dist.shape[1]
lon_range, lat_range = m2.makegrid(nx, ny)
x, y = m2(lon_range, lat_range)
# draw filled contours.
clevs = [0., 1., 1.5, 2.5, 3., 5, 7.5, 10, 15, 20, 30]
cs = m2.contourf(x,y,dist,clevs,cmap=cm.GMT_drywet, alpha=0.6, zorder=1)
m2.drawcoastlines(zorder=1)

plt.show()
```

```python
#calculate distances from OSM
#WARNING: TAKES HUGE AMOUNT OF TIME
#store data on disk

parsed = []
max_attempts = 5

if os.path.isfile('osm_dist_data.npy'):
    parsed = np.load('osm_dist_data.npy').tolist()
else:
    for lar in lat_range.reshape(11, 9):
        for lor in lon_range.reshape(11, 9):
            nearest_sts = []

            for i in range(len(lons)):
                nearest_sts.append((distance_on_unit_sphere(lar[4], lor[4], lats[i], lons[i]), i))
            nearest_sts.sort()
            for d,i in nearest_sts[0:5]:
                request = osm_api_url + 'loc=' + str(lats[i]) + ',' + str(lons[i])
                for la in lar:
                    for lo in lor:
                        request = request + '&loc=' + str(la) + ',' + str(lo)
                resp = requests.get(request)
                if resp.status_code == 400:
                    attempt = 0
                    while resp.status_code == 400:
                        if attempt == max_attempts:
                            resp.raise_for_status()
                        attempt += 1
                        sleep(600)
                        resp = requests.get(request)
                elif resp.status_code == 200:        
                    dist_resp = \
                        resp.text[resp.text.index('"distance_table":')+len('"distance_table":'):-1].encode('ascii','ignore')
                    dist_table = \
                        np.array(dist_resp[1:-1].replace('[','').replace(']','').split(','), dtype='int32').reshape(82,82)
                    parsed.append((lar, lor, i, dist_table))
                    sleep(45)
                else:
                    resp.raise_for_status()
    np.save('osm_dist_data', parsed)

parsed[0:1]
```

```python
grid = []
for k in [(l[0].tolist(), l[1].tolist()) for l in parsed]:
    for l in k[0]:
        for j in k[1]:
            grid.append((l,j))
grid = set(grid)
```
